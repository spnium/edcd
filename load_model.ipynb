```python
# Convert to TFLite format
converter = tf.lite.TFLiteConverter.from_keras_model(loaded_model)
tflite_model = converter.convert()
```

```python
# Save as a .tflite file
with open('eyeclss003.tflite', 'wb') as f:
    f.write(tflite_model)
```
