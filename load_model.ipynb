```python
import tensorflow as tf
```

```python
import tensorflowjs as tfjs
```

```python
loaded_model = tf.keras.models.load_model('./eyeclss003.keras')
loaded_model.summary()
```

```python
tfjs.converters.save_keras_model(loaded_model, "./tfjseyeclss003")
```

```python
import tf2onnx
import onnx
```

```python
input_signature = [tf.TensorSpec([None, 256, 256, 3], tf.uint8, name='x')]
onnx_model, _ = tf2onnx.convert.from_keras(loaded_model, input_signature, opset=13)
```

```python
onnx.save(onnx_model, "eyeclss003.onnx")
```
